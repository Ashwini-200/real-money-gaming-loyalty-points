# Should they base it on the amount of loyalty points? Should it be based on number of games? Or something else?

The company should base the bonus mainly on loyalty points, because these already combine deposits, withdrawals, and games played. But to keep long-term players engaged and attract new players to play more often, I recommend adding a small extra weight for number of games played.

This is because in a gaming app, retaining active players is more valuable than just collecting deposits — more games mean more platform activity, more fee income, and higher trust.

So, a fair way is:

80% of the bonus weight from loyalty points (covers spending & big players)

20% weight from number of games played (motivates regular play & engagement).

This keeps big spenders motivated but also rewards loyal daily players, balancing both revenue and player retention.

# Suggest a suitable way to divide the allocated money keeping in mind the following points:

```python
print("So as I explained above i will add the loyalty point based on 80% and 20%")

#80% of the bonus weight from loyalty points (covers spending & big players)

#20% weight from number of games played (motivates regular play & engagement).
```

```python
# Take top 50
top50 = ranking.head(50).copy()

# Use columns directly
total_points = top50['Loyalty_Point'].sum()
total_games = top50['Num_games_played'].sum()

# Compute share safely
top50['Points_Share'] = top50['Loyalty_Point'] / total_points
top50['Games_Share'] = top50['Num_games_played'] / total_games

# Combine 80% + 20%
top50['Final_Share'] = (0.8 * top50['Points_Share']) + (0.2 * top50['Games_Share'])

# Allocate ₹50,000
top50['Bonus_Amount'] = top50['Final_Share'] * 50000

top50['total_amount'] = top50['Final_Share'] + top50['Bonus_Amount']

# Show final result
print(top50[['User Id', 'Loyalty_Point', 'Num_games_played', 'Final_Share', 'Bonus_Amount', 'total_amount']])
```

# Part-C:-Is the current formula fair or unfair?


The current loyalty point formula is partly fair because it combines deposits, withdrawals, and games played. This means it rewards users for putting money in and playing. But from a business point of view, this is not fully balanced — for a gaming app, the main goal is to build an active player community, keep them loyal for the long term, and attract new players. Giving high weight to deposits and withdrawals does not fully support this goal — players who play often but deposit less may feel unrewarded.

# How to make it more robust?

Looking at how users really use the app, the loyalty point formula should:

* Give higher weight to games played and time spent playing

* Reward daily logins and active days in a month

* Add bonus points for inviting new players (to attract newcomers)

* Give small extra points for trying new features or game modes

* Keep some weight for deposits (cash flow matters), but reduce or remove withdrawal points

* Encourage players to stay active over a longer period, not just deposit once

* Also, improving the design and variety of the game will help keep users engaged — so rewarding players who try new updates can help too.
